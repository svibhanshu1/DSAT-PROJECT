## Introduction

In this project, the task is to create an ATM machine program which performs tasks of  depositing some amount, withdrawing some amount, transferring funds and checking account balance.

## Methodology

    The concepts of Object-oriented Programming(OOP) are used to achieve the end goal. 
    OOP mainly uses the idea of classes and objects. 

### What are classes in OOP?

    Classes can be defined as a user-defined data type. A class contains some attributes and methods. 
    Attributes can be called as the properties of the data type we are creating using a class. 
    A method is mainly a function which defines the behavior of a class.

### What are objects in OOP?

    An object is an instance of a class. It is a real world entity which has unique attributes and behavior.
    An object can also be called as an entity of type 'class'.
    
    
### Implementation
    
    In the ATM machine program, the saving balance and checking balance attributes have been initialized inside the constructor with a default value . The pin remains private as it is known only by the user. 
    
    Separate methods for checking account balance, depositing, withdrawing and transferring have been created and have been accessed at appropriate times . 



```python
class ATM :  
    # Defining a constructor.
    def __init__(self):
        self.checking_balance = 5000
        self.saving_balance = 2000

    # Login method with a pre-set PIN which I assume is already present in the bank server
    # It returns True if pin matches else it returns False
    def login(self, pin) :
        if pin == 1111:
            return True
        else:
            return False

    # Credit method to deposit funds
    def credit(self, amt):
        self.saving_balance += amt

    # Deposit method to withdraw funds
    def debit(self, amt):
        self.saving_balance -= amt

    # Display method to view balance in account
    def display(self):
        print("Current balance is "+str(self.saving_balance))

    # Transfer method to tranfer funds from checking balance to saving account
    def transfer(self, amt):
        self.checking_balance -= amt
        self.saving_balance += amt

obj = ATM() # an object of class ATM created

# User enters PIN twice
fp = obj.login(int(input("Enter Your PIN: "))) # Enter PIN
sp = obj.login(int(input("Please Re-enter Your PIN: "))) # Confirm PIN

if fp and sp:
    
    while True:

        Choice = int(input("""  
                Press 1 To Check Saving Balance
                Press 2 To Deposit Funds
                Press 3 To Withdraw Funds
                Press 4 To Transfer Funds
                Press 5 To Exit \n"""))

        if Choice == 1:
            obj.display()
            
        elif Choice == 2:
            obj.credit(int(input("Enter Amount To Deposit: ")))
            print("Successfull")
            obj.display()
            
        elif Choice == 3:
            amt=(int(input("Enter Amount To Withdraw: ")))
            
            if amt < obj.saving_balance:
                obj.debit(amt)
                print("Please Collect Your Cash")
                obj.display()
                
            else:
                print("Insufficient Funds")
                
        elif Choice == 4:
            amt=int(input("Enter Amount To Transfer: "))
            
            if amt < obj.checking_balance:
                obj.transfer(amt)
                print("Transcation Successfull")
                print("Current Checking Balance Is : ", obj.checking_balance)
                print("Current Saving Balance Is : ", obj.saving_balance)

            else:
                print("Insufficient Funds In Checking Account")
                
        elif Choice == 5:
            print("Thank You")
            print("Please Collect Your Card !!")
            break
            
        else:
            print("Invalid Choice")
            break
        
else:
    print("Wrong PIN")
```
