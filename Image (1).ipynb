```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
# Image Class
class Image():
    
    # Method for loading image
    def load(self):
        path = input('Enter the image path\n') # It takes input the path of image  
        self.img = io.imread(path)
        self.display()
    
    # Method for horizontal flipping the image
    def h_flip(self):
        print("Original Image")
        self.display()
        
        self.img = np.fliplr(self.img)
        
        print("Horizontal Flipped Image")
        self.display()

    # Method for vertical flipping the image
    def v_flip(self):
        
        print("Original Image")
        self.display()
        
        self.img = np.flipud(self.img)
        
        print("Vertical Flipped Image")
        self.display()
    
    # Method to invert the colors of the image
    def invert_color(self):
        
        print("Original Image")
        self.display()
        
        mask = np.full(self.img.shape, 255)
        self.img = mask - self.img
        self.img.astype(np.uint8)
        
        print("Inverted Image")
        self.display()

    
    # Method to convert image from RGB to Grayscale
    def grayscale(self):
        
        print("Original Image")
        self.display()
        
        self.img = np.dot(self.img[...,:3],[0.299,0.587,0.144])
        plt.imshow(self.img,cmap='gray')
        print("RGB 2 GRAY")
        plt.show()
      
    # Method to convert color image to black & white
    def black_white(self):
        
        print("Original Image")
        self.display()
        self.img = np.dot(self.img[..., ::],[255, 255, 255])
        plt.imshow(self.img, cmap='gray')
        print("B & W")
        plt.show()
    
    # Method to save file with default name as "NEW.JPG"
    def save(self):
        
        plt.imsave('new.jpg', self.img)
    
    # Method to display the image
    def display(self):
        
        plt.imshow(self.img)
        plt.show()
        
        
c=Image() # Object of Image Class

while True:
    
    choice = int(input(""" Enter your choice:
    1. Load Image
    2. Edit Image
    3. Save Image
    4. Quit """))
     
    if choice == 1:
        c.load()
        
        
    elif choice == 2:
        while True:
            
            opt = input(""" How do you want to edit your image?
            a. Horizontal Flip
            b. Vertical Flip
            c. Invert Colors
            d. Convert to greyscale
            e. Convert to B & W
            f. Exite """)


            if opt == 'a':
                c.h_flip()

            elif opt == 'b':
                c.v_flip()

            elif opt == 'c':
                c.invert_color()

            elif opt == 'd':
                c.grayscale()

            elif opt == 'e':
                c.black_white()
                
            elif opt=='f':
                break
            else:
                print("Invalid Input")
                break

    elif choice == 3:
        c.save()
        print("File Saved")

    elif choice == 4:
        break
        
    else:
        print("Invalid choice!!")



```
